```python
# computation of content cost function 

def content_cost(a_C, a_G):
    
    # retrieving the dims from a_G
    m, n_H, n_W, n_C = a_G.get_shape().as_list()
    
    # unrolling the shapes for multiplication
    a_C_unrolled = tf.reshape(a_C, shape=[m, n_H * n_W, n_C])
    a_G_unrolled = tf.reshape(a_G, shape=[m, n_H * n_W, n_C])
    
    # calculating the content cost
    J_content = 1/(4*n_H*n_W*n_C) * tf.reduce_sum((a_C_unrolled - a_G_unrolled)**2)
    
    return J_content
```

```python
# computation of gram matrix

def gram_matrix(A):
    
    # gram matrix is the dot product of a matrix with itself
    # The gram matrix calculates the correlation between different channels
    GA = tf.matmul(A, tf.transpose(A))
    
    return GA
```

```python
# computation of layer cost function

def layer_style_cost(a_S, a_G):
    
    m, n_H, n_W, n_C = a_G.get_shape().as_list()
    
    # getting the appropriate shapes for calculation
    a_S = tf.transpose(tf.reshape(a_S, shape=[n_H * n_W, n_C]))
    a_G = tf.transpose(tf.reshape(a_G, shape=[n_H * n_W, n_C]))
    
    # calculating the gram matrices of the S and G
    GS = gram_matrix(a_S)
    GG = gram_matrix(a_G)
    
    # calculating the layer style cost
    J_style_layer = 1/((2*n_H*n_W*n_C)**2) * tf.reduce_sum((GS - GG)**2)
    
    return J_style_layer 
```

```python
# assigning the layer style lambda values 

STYLE_LAYERS = [
    ('conv1_1', 0.2),
    ('conv2_1', 0.2),
    ('conv3_1', 0.2),
    ('conv4_1', 0.2),
    ('conv5_1', 0.2)]
```

```python
# computing style cost (layer cost for all the layers)

def style_cost(model, STYLE_LAYERS):
    
    J_style = 0
    
    for layer_name, coeff in STYLE_LAYERS:
        
        # extracting the tensors from our selected layers
        out = model[layer_name]
        a_S = sess.run(out)
        a_G = out
        
        # calculating the style cost for each layer
        J_style_layer = layer_style_cost(a_S, a_G)
        
        # calculating the total style cost 
        J_style += coeff * J_style_layer
        
    return J_style
```

```python
# computing the total cost 

def total_cost(J_content, J_style, aplha = 10, beta = 40):
    
    J = aplha*J_content + beta*J_style
    
    return J
```

```python
# resetting the graph 
tf.reset_default_graph()

# starting the interactive session
sess = tf.InteractiveSession()
```

```python
# save your content image in the "input content" folder
# rename it as you like and update the name below
# REQUIREMENT: the image should have width = 400 pixels and height = 300 pixels
content_image = scipy.misc.imread("input content/iut_resized3.jpg")
content_image = reshape_and_normalize_image(content_image)
#imshow(content_image)
```

```python
# save your style image in the "input style" folder
# rename it as you like and update the name below
# "Starry Night" has already been included 
# REQUIREMENT: the image should have width = 400 pixels and height = 300 pixels
style_image = scipy.misc.imread("input style/starry_night.jpg")
style_image = reshape_and_normalize_image(style_image)
```

```python
generated_image = generate_noise_image(content_image)
imshow(generated_image[0])
```

```python
# loading the pretrained model in this new session
model = load_vgg_model("pretrained-model/imagenet-vgg-verydeep-19.mat")
```

```python
# assigning the content image to the input for content cost computation
sess.run(model['input'].assign(content_image))

# choosing the layer conv4_2 to be the layer to compute content cost
out = model['conv4_2']
a_C = sess.run(out)
a_G = out

# computing the content cost
J_content = content_cost(a_C, a_G)
```

```python
# assigning the style image to the input for style cost computation
sess.run(model['input'].assign(style_image))

# computing the style cost
J_style = style_cost(model, STYLE_LAYERS)
```

```python
# compputing the total cost 
J = total_cost(J_content, J_style, aplha = 10, beta = 40)
```

```python
# defining the optimizer to be Adam with learning rate 2.0
optimizer = tf.train.AdamOptimizer(2.0)

# defining the training step
train_step = optimizer.minimize(J)
```

```python
# The number of iterations can be changed in the following line 
def model_nn(sess, input_image, num_iterations = 2000):
    
    # initializing the global variables
    sess.run(tf.global_variables_initializer())
    
    # assigning the input image to the model
    sess.run(model['input'].assign(input_image))
    
    for i in range(num_iterations):
        
        # running the train step
        sess.run(train_step)
        
        # updating the generated image pixel values every iteration
        generated_image = sess.run(model['input'])
        
        # Print every 20 iteration.
        if i%20 == 0:
            Jt, Jc, Js = sess.run([J, J_content, J_style])
            print("Iteration " + str(i) + " :")
            print("total cost = " + str(Jt))
            print("content cost = " + str(Jc))
            print("style cost = " + str(Js))
            
            # save current generated image in the "/output" directory
            save_image("output/" + str(i) + ".png", generated_image)
    
    # save last generated image
    save_image('output/generated_image.jpg', generated_image)
    
    return generated_image
```

```python
model_nn(sess, generated_image)
```
